```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

## fuctional method 

```python
plt.plot(x,y,'r',label='x')
plt.plot(y,x,'b',label='y')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('title')
plt.legend(loc=10)         #  loc=10 means center
plt.xlim(0,30)
plt.ylim(0,30)
plt.show() #not neccessary in jupyter notebook
```

```python
plt.plot(x,y)
plt.xlabel('x_label')
plt.ylabel('y_label')
plt.title("title")
plt.show()
```

## label

```python

plt.title('title',fontsize=50,color='blue',fontstyle='italic')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.grid()
plt.plot(x,y)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'m*',x,y,'r-')
plt.title('1')

plt.subplot(2,2,2)
plt.plot(y,x,'g*',y,x,'-')
plt.title('2')

plt.subplot(2,2,3)
plt.plot(x,y,'-.')
plt.title('3')

plt.subplot(2,2,4)
plt.plot(y,x,'.r')
plt.title('4')
```

## object oriented method

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('x_label')
axes.set_ylabel('y_label')
axes.set_title('title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3]) #[left,bottom,width,height]

axes1.plot(y,x)
axes1.set_title('larger plot')

axes2.plot(x,y)
axes2.set_title('smaller plot')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title("first")

axes[1].plot(y,x)
axes[1].set_title('second')

plt.tight_layout()     #doesnot overelap the graphs
```

## how to use legend 

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='x square')
ax.plot(x,x**3,label='x cube')

ax.legend(loc=0)  #we can use loc value from 0 to 10
# ax.legend(loc=(0.1,0.1))
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='red',linewidth=3,alpha=2,linestyle='-',marker='*',markersize=20,markerfacecolor='yellow',markeredgewidth=2,markeredgecolor='green')
```

## x,y limits

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,marker='*',markersize=23)

ax.set_xlim(0,3)
ax.set_ylim(0,5)
```

```python
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,0.2,0.2])

axes1.plot(x,y)
axes2.plot(x,y)
plt.show()
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,0.2,0.2])

axes1.plot(x,y,x,z)
axes2.plot(x,y,x,z)

axes1.set_xlim(0,100)
axes1.set_ylim(0,10000)
axes2.set_xlim(20,22)
axes2.set_ylim(300,500)
```

```python
fig,ax=plt.subplots(1,2,figsize=(9,2))
ax[0].plot(x,y,linewidth=4,linestyle='-')
ax[1].plot(x,z,linewidth=2,color='r',linestyle='--')
```
