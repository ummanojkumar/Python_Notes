## matrix plots

```python
tips=sns.load_dataset('tips')
flights=sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
c=tips.corr()
```

```python
sns.heatmap(c,annot=True,cmap='summer')
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidth=1)
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

## grids

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
sns.PairGrid(iris)
```

```python
import matplotlib.pyplot as plt
```

```python
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g= sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g.map(plt.scatter,'total_bill','tip')
```

## regression plots

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':30})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',hue='sex',aspect=0.6,height=8)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

## exersices

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic
```

```python
plt.figure(figsize=(12,1))
sns.set_style('whitegrid')
sns.set_context('poster',font_scale=0.5)
sns.jointplot(x='fare',y='age',data=titanic,joint_kws={'s':10})
```

```python
sns.distplot(titanic['fare'],kde=False,bins=40,color='red')
```

```python
titanic.head()
```

```python
sns.distplot(titanic["fare"],kde=False,bins=30) 
plt.ylim(0,500) 
plt.xlim(0,600)
```

```python
sns.boxplot(x='class',y='age',data=titanic)
```

```python
sns.swarmplot(x='class',y='age',data=titanic)
```

```python
sns.countplot(x='sex',data=titanic)
```

```python
s=titanic.corr()
```

```python
sns.heatmap(s)
plt.title('titanic')
```

```python
g= sns.FacetGrid(data=titanic,col='sex')
g.map(sns.distplot,'age',kde=False,bins=10)
plt.ylim(0,120)
plt.xlim(0,80)
sns.set_style('whitegrid')
```
